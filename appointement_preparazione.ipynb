```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## Lettura dei dati e stampa

```python
df = pd.read_csv("../Desktop/appointment_data.csv")
```

```python
y=df['No-show']
X=df.drop("No-show", axis=1)
```
